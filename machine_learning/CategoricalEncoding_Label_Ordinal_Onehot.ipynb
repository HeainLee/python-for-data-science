## **Categorical Encoding - 1 feature** ##

### **레이블인코딩 I/O , 변환 결과** ###

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = ['Male', 'Female', 'Female']
#X = np.array(['Male', 'Female', 'Female']).reshape(-1,1)
le.fit(X)
```

```python
print(le.classes_)
print(le.transform(X))
print(le.inverse_transform([1,0]))
```

### **오디널인코딩 I/O , 변환 결과** ###

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
#X = [['Male'], ['Female'], ['Female']]
X = np.array(['Male', 'Female', 'Female']).reshape(-1,1)
enc.fit(X)
```

```python
print(enc.categories_)
print(enc.transform(X)) # shape (3,1)
print(enc.inverse_transform([[1],[0]]))
```

### **원핫인코딩 I/O , 변환 결과** ###

```python
from sklearn.preprocessing import OneHotEncoder
oh = OneHotEncoder()
#X = [['Male'], ['Female'], ['Female']]
X = np.array(['Male', 'Female', 'Female']).reshape(-1,1)
oh.fit(X)
```

```python
print(oh.categories_)
print(oh.transform(X).toarray()) #shape (3.2) 
print(oh.inverse_transform([[0, 1], [1, 0]]))
```

```python
def inspect_condition(params_str): 
    params_dict = {}
    condition_list = params_str.split(', ')
    if len(condition_list) != 1: #요청한 파라미터 변경 개수가 2개 이상인 경우 
        for request_conidtion in condition_list:
            key_, values_ = request_conidtion.split(':')[0], request_conidtion.split(':')[1]
            params_dict[key_] = values_
    else: #요청한 파라미터 변경 개수가 하나인 경우 
        params_dict[condition_list[0].split(':')[0]] = condition_list[0].split(':')[1]
    return params_dict
```
